#Multuple boolean expressions
Remember for an expression that has (and)all the outputs will be false unless both the choices are true

For another expression that has (or)all the outputs are true if one choice is true and will read false only if both the choices are false

```python
a=1
b=2
c=3
print(a==b and b==c)
print(a<b or b>c)
print(a<b and b>c)
```

write a program that takes a value from the user. the program should print

.divisible by two if the value is divisible by two

.divisible by three if the value is divisible by three

.divisible by both two and three if the value is divisible by both two and three

otherwise it should print not divisibe by both two and three

```python
number=int(input("Enter a number"))
if number == (f"number%2 is 0"):
  print("divisible by two")
elif number == (f"number%3 is 0"):
  print("divisible by three")
elif number == (f"number%2 and 3 is 0"):
  print("divisible by both two and three")
else:
  print("not divisible by both two and three")
```
