#Loops
Allows for repetition of a word/ task

There are two types of loops

While loop

For loop

#While loop
when writing a while loop ensure that you have that control statement to prevent it looping for infinity times

```python
#simple program that prints hello world 5 times
Counter=5 #holds the number of iterations
while Counter>0:
  print("hello world")
  Counter=Counter-1
```

```python
#other example to make it repeat 5 times
counter=1
while counter<=5:
  print("hello world")
  counter=counter+1
```

```python
counter=1
while counter<4:
  print("value of counter",counter)
  print("***"*counter)
  counter=counter+1
  print("value of counter",counter)
```

```python
is_correct=True
password="secret"
while is_correct:
  user_pass=input("Enter your password")
if user_pass==password
  print("welcome back")
  is_correct=False
else:
  print("incorrect password!! try again")
```
