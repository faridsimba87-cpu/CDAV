#Loops

There are two types of loops

#While loop
when writing a while loop ensure that you have that control statement to prevent it looping for infinity times

```python
#simple program that prints hello world 5 times
Counter=5 #holds the number of iterations
while Counter>0:
  print("hello world")
  Counter=Counter-1
```

```python
#other example to make it repeat 5 times
counter=1
while counter<=5:
  print("hello world")
  counter=counter+1
```

```python
counter=1
while counter<4:
  print("value of counter",counter)
  print("***"*counter)
  counter=counter+1
  print("value of counter",counter)
```

```python
password_incorrect = True
password = "secret"

while password_incorrect:
    user_pass = input("Enter your password: ")

    if user_pass == password:
        print("Welcome back")
        password_incorrect = False
    else:
        print("Incorrect password. Try again!")
```

```python
counter = 7

while counter > 0:
    spaces = 7 - counter
    stars = 2 * counter - 1
    print(" " * spaces + "*" * stars)
    counter -= 1
```
