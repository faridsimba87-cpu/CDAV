make a program where the user enters any amount of integers and uses the =(equals sign) to end the process
the program should ensure that the user is able to find the lowest value, the sum, the highest value and the average
ensure that even if the user enters sth other than a value it does not collpase but prompts the user to enter an integer

```python
values_list = []
while True:
  value_str = input("Enter an integer (or = to finish): ")
  if value_str == "=": # Check for the termination character directly on value_str
    break
  values_list.append(float(value_str))

if values_list:
  total_sum = sum(values_list)
  count = len(values_list)
  average = total_sum / count
  maximum = max(values_list)
  minimum = min(values_list)
  print(f"The sum of the values is {total_sum}")
  print(f"The average of the values is {average}")
  print(f"The maximum value is {maximum}")
  print(f"The minimum value is {minimum}")
else:
  print("No values were entered.")
```

#List iterations with loops
if u want to print the list foing downwars without using indexing u use the for loop

```python
numbers=[12,13,14,15,16,17,18]
for num in numbers:
  print(num)
  high=max(numbers)
  low=min(numbers)
  print("the highest number is",high)
  print("the lowest number is",low)
```

```python
numbers=[13,11,16,15,16,17,18]
highest=numbers[0]
for num in numbers:
  if num>highest:
    highest=num
print(highest)
```

```python
student=[12,16,19,24,10,15,18,20]
count=0
for grade in student:
  if grade>=15:
    count+=1
print(count)

```

#list comprehension
In Python, List Comprehensions provide a shorter, highly
readable syntax to create a new list from an existing list

```python
even=[]
evens = [x for x in range(1,51) if x % 2 == 0]
even.append(evens)
print(evens)
```
