### arguements (think of this as variables)
are basically variables initialised within the brackets of a function definition and used within the function
###parameters (think of this as values)
are values passed when calling the function within the brackets, they correspond with the arguments present when the function was created


```python
#addition function
#the a and b is like in the context of algebra where x and y rep sth
def add_3(a,b):
  return a+b
```

```python
add_3(50,25)
```

```python
#the lang is used as a variable which is used as an argument and is used to rep the key at the dict
def greeter_3(lang):
  greetings={"fr":"bonjour",
             "sw":"jambo",
             "en-br":"hi",
             "jp":"konichiwa"
             }
  return greetings[lang]
```

```python
print(f"{greeter_3('jp')} farid")
```

```python
def sun_diff(x,y):
  s= x+y
  d= x-y
  return (s,d)
```

```python
sum_of_both, diff_of_both= sun_diff(10,20)
print(sum_of_both)
print(diff_of_both)
```
