#For Loop
For loops handles variables incementation automatically behid the scenes, eliminating the risk of infinite loops

```python
for number in range(5):
  print(number)
```

more examples
the last two below is used to show the incemental value

```python
for i in range(0,11):
  for j in range(0,11):
    print(f"{i} x {j} = {i*j}")
  print()
```

```python
CELL_WIDTH = 4

def print_horizontal_line(num_elements):
    """Prints a horizontal line for the table based on the number of elements."""
    print("+" + ("-" * CELL_WIDTH + "+") * num_elements)

# The table will have 1 header column ('X') and 10 number columns (1-10).
num_table_columns = 11

# Print the top border
print_horizontal_line(num_table_columns)

# Print the header row (X and numbers 1-10)
header_row = "|{:^{width}}".format("X", width=CELL_WIDTH)
for j in range(1, 11):
    header_row += "|{:^{width}}".format(j, width=CELL_WIDTH)
header_row += "|"
print(header_row)

# Print the separator line after the header
print_horizontal_line(num_table_columns)

# Print the multiplication table rows
for i in range(1, 11):
    row_str = "|{:>{width}}".format(i, width=CELL_WIDTH) # Row header (multiplier)
    for j in range(1, 11):
        product = i * j
        row_str += "|{:>{width}}".format(product, width=CELL_WIDTH)
    row_str += "|"
    print(row_str)

# Print the bottom border
print_horizontal_line(num_table_columns)
```

```python
print("\t" , end="")
#column headings
for i in range(1,11):
  print(i, end="\t")

print()

#primt rows
for i in range(1,11):
  print(i, end="\t")
  for j in range(1,11):
    print(i*j, end="\t")
  print()
```

#The break keyword


```python
for number in range(10):
  if number == 5:
    break
  print(number)
```

#The continue Keyword
the continue word is used to tell the python to skip the number stated by the if statement

```python
for number in range(10):
  if number==5:
    continue
  print(number)
```
